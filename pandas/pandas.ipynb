Split->Split the DataFrame into groups.
Aplly->Perform an operation on each group.
Combine->Combine the results into a new object.

```python
df2 = pd.DataFrame({
    "Name": ["Alice", "Bob", "Charlie", "David", "Eva"],
    "Department": ["IT", "HR", "IT", "HR", "IT"],
    "Salary": [70000, 50000, 90000, 55000, 80000],
    "Age": [22, 25, 30, 28, 35]
})

print(df2)
```

```python
#group by department or split by department 
groups = df2.groupby("Department")
print(groups)
```

```python
# Apply any operation on it  and combine it 
df2.groupby("Department")["Salary"].mean()
```

Aggregate Multiple Coloumn With Multiple Function(Grouped)

```python
result = df2.groupby("Department").agg({
    "Salary": ["sum", "mean", "max"],
    "Age": ["min", "max", "mean"]
})

result
```


merge()	-> Combine DataFrames based on common columns or keys--JOIN
concat() -> Stack DataFrames vertically or horizontally. --	UNION / column binding
join()	-> Join DataFrames using their indexes (or a key) --	JOIN

```python
df3 = pd.DataFrame({
    "ID": [1, 2, 3],
    "Name": ["Alice", "Bob", "Charlie"]
})

df4 = pd.DataFrame({
    "ID": [1, 2, 4],
    "Salary": [50000, 60000, 70000]
})

result = pd.merge(df3, df4, on="ID", how="inner")
print(result)
```

Merge Types 

```python
pd.merge(df3, df4, on="ID", how="inner")   # Common rows
pd.merge(df3, df4, on="ID", how="left")    # All rows from left
pd.merge(df3, df4, on="ID", how="right")   # All rows from right
pd.merge(df3, df4, on="ID", how="outer")   # All rows from both
```
