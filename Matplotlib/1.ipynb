```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
x = np.linspace(0,5,11)
x
y = x**2
y
```

```python
#plot generates the line graph 
plt.plot(x,y)
# x axis and y are axis 
# lables and legends and title 
plt.title('starting graph')
plt.xlabel("X axis")
plt.ylabel("Y axis")
```
