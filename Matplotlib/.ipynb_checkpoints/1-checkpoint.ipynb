```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
x = np.linspace(0,5,11)
x
y = x**2
y
```

```python
#plot generates the line graph 
plt.plot(x,y,color = "red",linestyle="--", marker="o")
# x axis and y are axis 
# lables and legends and title 
plt.title('starting graph')
plt.xlabel("X axis")
plt.ylabel("Y axis")
```

```python
# sub plot -> Matlab Multiple plot ik saath bana na
#+-------------------------------+
#|           Figure              |
#|                               |
#| +---------+  +---------+      |
#| | Graph 1 |  | Graph 2 |      |
#| +---------+  +---------+      |
#|                               |
#| +---------+  +---------+      |
#| | Graph 3 |  | Graph 4 |      |
#| +---------+  +---------+      |
#+-------------------------------+

plt.subplot(2,2,1)
plt.subplot(2,2,2)
plt.subplot(2,2,3)
plt.subplot(2,2,4)
plt.plot(x,y)# for specific 4th one 
```

```python
#object oriented way and white background is called canvas
fig = plt.figure()
fig.add_axes([0.1,0.2,0.3,0.4])
plt.figure(figsize=(6,7))

plt.figure(figsize=(6,4),dpi=100)

x = [1,2,3,4]
y = [10,20,15,25]

plt.plot(x, y)

plt.show()
```

```python
#plt.grid()-> Graph mein grid (horizontal aur vertical lines) dikhata hai.
```

```python
#legend ->Jab ek se zyada lines hoti hain, tab unka naam dikhata hai.
x = [1,2,3,4]

sales = [10,20,15,25]

plt.plot(x, sales, label="Sales")

plt.legend()

plt.show()
```

```python
x = [1,2,3,4]

sales = [10,20,15,25]
profit = [5,10,8,15]

plt.plot(x, sales, label="Sales")
plt.plot(x, profit, label="Profit")

plt.title("Company Report")
plt.xlabel("Months")
plt.ylabel("Amount")

plt.grid(True)

plt.legend()

plt.show()
```
