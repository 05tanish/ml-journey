```python
import numpy as np 
```

```python
#create arrays 

arr1 = np.array([1,2,3,4,5])
b = np.zeros((2,3))
c = np.ones((2,3))
d = np.arange(0, 10, 2)       # Values from 0 to 8 with step 2
e = np.linspace(0,3,7)
f = np.eye(3,3) #identify matrix
g = np.random.rand(2,4) #random values between 0 and 1

print("Array 1:", arr1)
print("Array 2:", b)
print("Array 3:", c)
print("Array 4:", d)
print("Array 5:", e)
print("Array 6:", f)
print("Array 7:", g)
```

```python
# Array Attributes

arr2 = np.array([[1,2,3],[5,6,9]])

print("arr2size:",arr2.size)
print("arr2shape:",arr2.shape)
print("arr2dimensions:",arr2.ndim)
print("arr2datatypes",arr2.dtype)
#shape → Number of rows and columns.
#ndim → Number of dimensions.
#size → Total number of elements.
#dtype → Data type of elements.
```

```python
#Indexing and Slicing
arr3 = np.array([10,20,30,40,50])
print("arr 3rd element:", arr3[2])
print("arr range of elements:",arr3[1:4]) #from index 1 to 3
print("arr last element:",arr3[-1])
print("every 3rd element:",arr3[::3]) #from start to end with step 3
```

```python
#2D Indexing
arr4 = np.array([[1,2,3],
                [4,5,6],[7,2,5]])
print("specific element of 2d array:",arr4[1,2])
print("complete column of 2d array :",arr4[:,2])
print("complete row of 2d array :",arr4[1,:])
```

```python
#Reshaping Arrays

arr5 = np.arange(12)
print(arr5)
arr_b = arr5.reshape(3,4)
print(arr_b)
arr_c = arr5.reshape(2,6)
print(arr_c)

```

```python
arr_d =arr5.flatten()
print(arr_d) #flatten() returns a copy.
arr_e = arr5.ravel()
print(arr_e)
#ravel() returns a flattened view of the array. It does not create a copy of the original array. 
# If you modify the raveled array, it will affect the
```

```python
#Mathematical Operations
a = np.array([1,2,3])
b = np.array([4,5,6])

print(a + b)
print(a - b)
print(a * b)
print(a / b)
print(a ** 2)
```

```python
#Universal Functions (ufuncs)
arr6 = np.array([2,3,4])
print(np.sqrt(arr6))
print(np.exp(arr6))
print(np.log(arr6))
print(np.sin(arr6))
print(np.cos(arr6))
```

```python
# Aggregation Functions
arr7 = np.array([[1,2,3,4,5],[5,7,82,4,7]])
print(np.sum(arr7))
print(np.sum(arr7, axis=1)) #axis 1 means sum along the rows, axis 0 means sum along the columns    
print(np.sum(arr7, axis=0))

print(np.mean(arr7))
print(np.median(arr7))
print(np.std(arr7))
print(np.max(arr7))
print(np.min(arr7))
print(np.var(arr7))
```

```python
#Broadcasting
arr8 = np.array([1,2,3])
print(arr8 + 5) #broadcasting adds 5 to each element of the array

arr9 = np.array([2,5,8])
arr10 = np.array([4,6,9])
arr11 = arr9+arr10
print(arr11)
```

```python
#Boolean Masking
arr12 = np.array([1,2,5,64,55,8,96])
print(arr12 > 5)
print(arr12 >= 5)
print(arr12[arr12>5])

print(arr12 [arr12 % 2 == 0])
print(arr12[(arr12 >= 5) & (arr12 <= 10)])
print(arr12[(arr12 >= 5) | (arr12 <= 10)])
arr12[arr12 > 30] = 100
print(arr12)
```
